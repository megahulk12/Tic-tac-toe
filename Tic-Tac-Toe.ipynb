```python
def play_again():
    choice = 'meh'
    
    while choice not in ['Y', 'N']:
        choice = input('Play again? Y or N: ')
        
        if choice not in ['Y','N']:
            print('Invalid syntax\n Y or N?')
            
    if choice == 'Y':
        return True
    elif choice == 'N':
        return False
```
